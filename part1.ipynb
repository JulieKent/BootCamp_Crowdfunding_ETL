# Create Category and Subcategory DataFrames

```python
import pandas as pd
```

```python
# Extract and transform the excel data to create a category DataFrame 

campaign = pd.read_excel("Resources/crowdfunding.xlsx")
campaign.head()
```

```python
# Seperate the column "category & sub-category" values

category_df = campaign['category & sub-category'].str.split('/', expand=True)
category_df.columns = ['category', 'subcategory']

category_df
```

```python
# Extract unique categories and subcategories
unique_categories = category_df['category'].unique()
unique_subcategories = category_df['subcategory'].unique()
```

```python
# Create a category DataFrame with the "category_id" column
category_df = pd.DataFrame({"category_id": [f"cat{i+1}" for i in range(len(unique_categories))],
                            "category_name": unique_categories})

category_df
```

```python
# Export the category DataFrame to a CSV file
category_df.to_csv("Outputs//category.csv", index=False)
```

```python
# Create a subcategory DataFrame with the "subcategory_id" column
subcategory_df = pd.DataFrame({"subcategory_id": [f"subcat{i+1}" for i in range(len(unique_subcategories))],
                            "subcategory_name": unique_subcategories})
```

```python
# Export the subcategory DataFrame to a CSV file
subcategory_df.to_csv("Outputs//subcategory.csv", index=False)
```
